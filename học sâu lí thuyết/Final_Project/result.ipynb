```python
with open("saved_models\machinelearning_model.pickle", "rb") as file:
    machinelearning_model = pickle.load(file)
input_data = np.array([0, 1, 1, 1, 1, 1, 10, 2, 3, 10, 10, 1200, 0.9, 2, 17, 20, 50, 3, 100, 1800, 15, 200, 2, 1, 1])
input_data = input_data.reshape(1, -1)
machinelearning_model = round(machinelearning_model.predict(input_data)[0], 2)
```

```python
name_of_model = ["model có 10 lần epoch và batchsize 16",
    "model có 10 lần epoch và batchsize 32",
    "model có 10 lần epoch và batchsize 64",
    "model có 10 lần epoch và batchsize 128",
    "model có 20 lần epoch và batchsize 16",
    "model có 20 lần epoch và batchsize 32",
    "model có 20 lần epoch và batchsize 64",
    "model có 20 lần epoch và batchsize 128",
    "model có 50 lần epoch và batchsize 16",
    "model có 50 lần epoch và batchsize 32",
    "model có 50 lần epoch và batchsize 64",
    "model có 50 lần epoch và batchsize 128",
    "model có 100 lần epoch và batchsize 16",
    "model có 100 lần epoch và batchsize 32",
    "model có 100 lần epoch và batchsize 64",
    "model có 100 lần epoch và batchsize 128",]
scores = [model_with_10epochs_batchsize_16, 
    model_with_10epochs_batchsize_32, 
    model_with_10epochs_batchsize_64, 
    model_with_10epochs_batchsize_128,
    model_with_20epochs_batchsize_16,
    model_with_20epochs_batchsize_32,
    model_with_20epochs_batchsize_64,
    model_with_20epochs_batchsize_128,
    model_with_50epochs_batchsize_16,
    model_with_50epochs_batchsize_32,
    model_with_50epochs_batchsize_64,
    model_with_50epochs_batchsize_128,
    model_with_100epochs_batchsize_16,
    model_with_100epochs_batchsize_32,
    model_with_100epochs_batchsize_64,
    model_with_100epochs_batchsize_128,]
machinelearning_score = machinelearning_model
```

```python
def print_result():
    print("Kết quả mong muốn trong khoảng từ 3.6 - 4.0")
    print(f"Kết quả dự đoán từ model Machine Learning: {machinelearning_score}")
    for i in range(len(scores)):
        print("Kết quả nhận được từ", name_of_model[i], ":", scores[i])
```

```python
def calculate_acc():
    for score in range(len(scores)):
        accuracy = round((scores[score] / machinelearning_score) * 100, 2)
        if accuracy < 100:
            pass
        else:
            accuracy = round((machinelearning_score /scores[score] ) * 100, 2)
        print("Độ chính xác của", name_of_model[score], f"so với model Machine learning là: {accuracy}%")
```

```python
print_result()
```

```python
calculate_acc()
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(10, 8))
plt.barh(range(len(scores)), scores)
plt.yticks(range(len(scores)), name_of_model)
plt.axvline(0, color='black', linestyle='--')  # Add a vertical line at the middle
plt.xticks([-100, -75, -50, -25, 0, 25, 50, 75, 100])  # Adjust X-axis ticks
plt.xlim(-100, 100)
plt.ylabel('Model')
plt.xlabel('Accuracy')
plt.title('Accuracy of DL Models Versus Machine Learning Model')

# Adjust the layout to center the plot
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_linewidth(0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
